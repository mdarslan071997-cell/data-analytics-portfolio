```python
import pandas as pd
```

```python
# loading dataset
df = pd.read_csv("SampleSuperstore.csv")
# checking first few rows
df.head()
```

```python
# checking shape
df.shape
```

```python
# checking data types
df.info()
```

```python
# summary statistics
df.describe()
```

```python
# Checking null value
df.isnull().sum()
```

```python
# To Find Out Columns
print(df.columns)
```

```python
# Total Business Performance
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_discount = df["Discount"].mean()
total_orders = len(df)
```

```python
# Total Sales
print("total_sales:",round(total_sales,2))
```

```python
# Total Profit
print("total_profit:",round(total_profit,2))
```

```python
# Total Discount
print("total_discount:",round(total_discount,2))
```

```python
# Total Orders
print("total_orders:",total_orders)
```

```python
# Sales by Category

category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(category_sales)
```

```python
#Profit by Region

profit_region = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

print(profit_region)
```

```python
# Identify Loss-Making Sub-Categories

sub_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)
df.head(10)
```

```python
# Identify Loss-Making Sub-Categories

sub_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)
df.head
```

```python
# Discount Analysis without after mean bracket
discount_profit = df.groupby("Discount")["Profit"].mean
print(discount_profit)
```

```python
# Discount Analysis
discount_profit = df.groupby("Discount")["Profit"].mean()
print(discount_profit)
```

```python
# Profit Margin Column

df["Profit_Margin"] = df["Profit"] / df["Sales"]
df.groupby("Category")["Profit_Margin"].mean()
```
